```python
from scipy.integrate import ode
from numpy import sign, array, tanh
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def FrictionSpring(t, y):
    mu = 0.1
    km = 2
    g = 9.8
    x = y[0] # Remind ourselves of the convention in y
    v = y[1]
    return array([v, -km*x - mu*sign(v)])
```

```python
y0, t0 = [2.0, 0.0], 0


r = ode(FrictionSpring)
#r = ode(f).set_integrator('zvode', method='bdf')
#r = ode(f).set_integrator('dopri5', method='bdf')


r.set_initial_value(y0, t0)

t1 = 40
dt = 0.1
times=[]
ys=[]
while r.successful() and r.t < t1:
    times.append(r.t)
    ys.append(r.integrate(r.t+dt))


plt.plot(times,ys)
```
