```python
from scipy.integrate import odeint
from numpy import sign, array, tanh, linspace
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def FrictionSpring(y, t):
    mu = 0.05
    km = 2
    g = 9.8
    x = y[0] # Remind ourselves of the convention in y
    v = y[1]
    return array([v, -km*x - mu*sign(v)])
```

```python
N = 1000
y = array([1.0,0.0]) # initial y values
t = linspace(0,40,N)

answer = odeint(FrictionSpring, y, t)

plt.plot(t,answer)
```
