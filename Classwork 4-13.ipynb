```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 500
I = 100
y=np.zeros([I,N])
x = np.linspace(-5,5,I)
```

```python
y[:,0]=np.cos(np.pi*x/10)
#y[:,0]=np.exp(-(x-2)**2)
y[:,1]=y[:,0]
r=1
```

```python
for n in range (1,N-1):    
    for i in range (1,I-1):
        y[i,n+1]=2*(1-r**2)*y[i,n]-y[i,n-1] + r**2 * (y[i+1,n]+y[i-1,n])
        
```

```python
plt.imshow(y)
```
