```python
import numpy as np
import numpy.random as nprand
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
def Pcycle (state,time,gamma,bet,alpha):
    g0 = bet*state[1]*state[0]-gamma*state[0]
    g1 = alpha*state[2]-bet*state[1]*state[0]
    g2 = gamma*state[0]-alpha*state[2]
    return np.array([g0,g1,g2])
```

```python
x1=0.2    
x2=0.12
x3=1.0
gamma = 1/4
alpha=1/20
bet=1/0.4


t=np.linspace(0,100,10000)
answer = odeint(Pcycle,[x1,x2,x3],t,args=(gamma,bet,alpha))

plt.plot(t,answer[:,0:2])
```
