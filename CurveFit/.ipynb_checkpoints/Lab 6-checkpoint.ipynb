```python
#4

def func2(x,a,b,c):
    y = a*(np.sin(x)/c)+np.cos(x)*x**2-(x*b)
    return y
```

```python
a=2
b=4
c=.5

x=np.linspace(0,40,500)

ydata3 = func2(x,a,b,c) * (1.0 + 0.5*np.random.normal(size=len(x)))
```

```python
plt.plot(x,func2(x,a,b,c),".")
```

```python
plt.plot(x,ydata3,".")
```

```python
parameters3, covariance = curve_fit(func2, x, ydata3)
parameters3
```

```python
covariance
```

```python
yfit2 = func2(x, *parameters3) 
```

```python
plt.plot(x,ydata3,".")
plt.plot(x,yfit2)
plt.plot(x,func2(x,a,b,c))
```

```python
yi=func2(x,a,b,c)

plt.plot(x,((yfit2-yi)/yi)*100)
plt.title("Fit error%")
```

```python
#5


def DrivenDampedPendulum(state, timep, betap, w_0, gammap, wd):

    xdot=state[1]

    vdot=-((w_0)**2)*np.sin(state[0])-2*(betap)*state[1]+(gammap)*(w_0**2)*np.cos(wd*timep)

    return array([xdot,vdot])
```

```python
%pylab inline
from scipy.integrate import odeint
```

```python

wd=2*np.pi  
w_0=1.5*wd  
beta = w_0/4
theta_0=0.0  
v_theta_0 = 0.0
gamma=0.2

t_max=.24
N=500

time=np.linspace(0,t_max,N)


solution=odeint(DrivenDampedPendulum, [theta_0,v_theta_0], time, args=(beta, w_0, gamma, wd))


figure()
plot(time,solution[:,0]/(2*np.pi),"k-")
xlabel("time (drive periods)")
ylabel("angle")
xlim(0,t_max)
ax1=gca()
```

```python
ydata4 = solution[:,0]/(2*np.pi) * (1.0 + 0.2*np.random.normal(size=len(x)))
```

```python
plt.plot(x,ydata4,".")
```

```python
parameters, covariance = curve_fit(solution[:,0]/(2*np.pi), x, ydata4)
parameters
```
